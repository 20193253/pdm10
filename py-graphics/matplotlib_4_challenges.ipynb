> [hint] 검색: "How to add title to subplots in Matplotlib?"

```
import numpy as np
import matplotlib.pyplot as plt
```

```
plt.hist(np.random.rand(1000))
```

```
plt.hist(np.random.randn(1000))
```

```
plt.figure(figsize=(10, 5))
plt.subplot(121).set_title("Uniform distribution")
plt.xlabel('x') 
plt.ylabel('y') 
plt.hist(np.random.rand(1000))

plt.subplot(122).set_title("Normal distribution") 
plt.xlabel('x') 
# plt.ylabel('y') 
plt.hist(np.random.randn(1000))

plt.suptitle("Uniform distribution & Normal distribution", fontweight="bold")
plt.show()
```

### [도전 2.] 0~9 사이의 무작위 정수 1000개를 만들어서 각 수의 빈도에 대한 히스토그램, 막대그래프, 꺽은선 그래프를 그리고 비교하시오.
- subplot 이용
- style(제목, x-축 레이블, y-축 레이블, ...)



```
n1000 = np.random.randint(0, 10, 1000)
```

```
value, count= np.unique(n1000, return_counts=True)
# count
```

```
plt.hist(n1000)
plt.show()
```

```
plt.bar(value,count,color="bgrcmyk")  # 'bgrcmykw
```

```
plt.plot(value, count, marker = 'o')
plt.show()
```

```
plt.figure(figsize=(15,5))

plt.subplot(131).set_title("Histogram") 
plt.xlabel('number') 
plt.ylabel('count')
plt.hist(n1000)

plt.subplot(132).set_title("Bar Graph") 
plt.xlabel('number')
# plt.ylabel('count')
plt.bar(value,count,color="bgrcm")

plt.subplot(133).set_title("Line Graph")
plt.xlabel('number')
# plt.ylabel('count')
plt.plot(value,count,marker='o')

plt.suptitle("Histogram & Bar Graph & Line Grapph", fontweight="bold")
plt.show()
```
